#### 4.Find out under which team and position, spending in terms of salary is high.

```python
df["Salary"]
```

```python
u_positions = df["Position"].unique()

for position in u_positions:
    filtered_df = df[df["Position"] == position][["Name","Position","Salary"]]
    print(f"Employees with position '{position}':")
    print(filtered_df)
    total_salary = filtered_df["Salary"].sum()
    print(f"Total Salary for '{position}': {total_salary}")
    print("\n")
```

```python
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")
sns.barplot(x='Position', y='Salary', data=df, estimator=sum, ci=None, palette='viridis')
plt.xlabel('Position')
plt.ylabel('Total Salary')
plt.title('Total Salary by Position')
```

```python
u_team = df["Team"].unique()

for team in u_team:
    filtered_df1 = df[df["Team"] == team][["Name", "Team", "Salary"]]
    print(f"Employees with team '{team}':")
    total_salary = filtered_df1["Salary"].sum()
    print(f"Total Salary for '{team}': {total_salary}")
    print("\n")
```

```python
plt.figure(figsize=(12, 6))
plt.xticks(rotation=90)
sns.set_theme(style="whitegrid")
sns.barplot(x='Team', y='Salary', data=df, estimator=sum, ci=None, palette='plasma')
plt.xlabel('Teams')
plt.ylabel('Total Salary')
plt.title('Total Salary by Team')
```

#### 5.Find if there is any correlation between age and salary , represent it visually.

```python
df2=df[['Age','Salary']]
df2
```

```python
corr = df2.corr()
print(corr)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix between Age and Salary")
plt.show()
```

##### INFERENCE
- The correlation analysis conducted on the employee dataset reveals insightful patterns between the variables "Age" and "Salary." Examining the correlation matrix, the diagonal entry of 1.00000 for "Age vs. Age" is an expected perfect positive correlation, representing a variable's correlation with itself.

- Moving to the key focus of interest, the correlation coefficient of 0.21117 between "Age" and "Salary" indicates a positive but relatively weak relationship. In simpler terms, as employees' ages increase, there is a tendency for their salaries to experience a modest rise. This finding is corroborated by the symmetric nature of correlation, as evidenced by the mirrored value of 0.21117 for "Salary vs. Age." The overall interpretation suggests a positive yet not particularly robust association between age and salary in the employee dataset.
